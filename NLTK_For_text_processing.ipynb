```python
import nltk
from nltk.stem import PorterStemmer # Used for stemming
from nltk.corpus import stopwords
```

# `Text Preprocessing`

### `Step :- 1  Tokenization`

```python
# TODO: Tokenization 1 (Converts paragraph into individual Sentences then focus on individual words)
nltk.download('punkt')  # Download the necessary resources for tokenization
senteces = nltk.sent_tokenize(paragraph)  # Converts paragraph into individual sentences
```

### `Stemming`

```python
# Heling to find out the base or root word of the word
# example: loved, loving, loves => lov3
```

```python
stemmer = PorterStemmer()
```

```python
stemmer.stem("Loved")  # Converting into base words
```

```python
stemmer.stem("thinking")  # Converting into base words
```

```python
stemmer.stem("drink")  # Converting into base words
```

### `Lametization`

```python
# Tokenization :- For converting paragraph into individual correct words
```

```python
from nltk.stem import WordNetLemmatizer    # For Lammetization
lemmatizer = WordNetLemmatizer()
```

```python
lemmatizer.lemmatize("goes")  # Converting into base words
```

### `Cleaning Text`

```python
len(senteces)
```

```python
# TODO: Applying Regular Expression such that removing special characters
import re
corpus = []
for i in range(len(senteces)):

    # This says what to do with the senteces, it says that if 
    # there is any character other than a-z or A-Z then replace it with space
    review = re.sub('[^a-zA-Z]', ' ', senteces[i])  
    review = review.lower()  # Converting into lower case
    # review = review.split()  # Converting into list of words
    corpus.append(review)    # Appending the list of words into corpus
```

```python
# This is the final output of the corpus which is a list of list of words 
# corpus
```

### `Appllying Stemming For Every Word`

```python
for i in corpus:
    # It will convert the list of words into string and then tokenize it into individual words
    words = nltk.word_tokenize(str(i))  # Converts sentences into words
    for word in words:
        # Stemming the words if they are not stop words(common words in english language)
        if word not in set(stopwords.words('english')):
            print(stemmer.stem(word))
```

### `Applying Lammetizer`

```python
# It displays the stop words in english language
# stopwords.words('english')
```

```python
for i in corpus:
    words = nltk.word_tokenize(str(i))
    for word in words:
        # Stemming the words if they are not stop words(common words in english language)
        if word not in set(stopwords.words('english')):
            print(lemmatizer.lemmatize(word))
```

# `Vectorization`

### `Bag Of Words(BOW)`

```python
# CountVectorizer :- It is a tool that converts a collection of text documents 
# to a matrix of token counts. It is used to create a bag-of-words representation of the text data, 
# which can then be used for various natural language processing tasks such as text classification, 
# clustering, and topic modeling.
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()  # binary=True means it will only consider the presence or absence of a word, not the frequency,
```

```python
# Applying stopwords to the CountVectorizer
import re
corpus = []
for i in range(len(senteces)):
    review = re.sub('[^a-zA-Z]', ' ', senteces[i])
    review = review.lower()
    review = review.split()
    review = [word for word in review if word not in set(stopwords.words('english'))]
    review = ' '.join(review)
    corpus.append(review)
```

```python
X = cv.fit_transform(corpus)
```

```python
# It will display the unique words in the corpus and their corresponding index in the matrix
cv.vocabulary_
```

```python
# 
corpus[0]
```

```python
X[0].toarray()  # It will display the count of each word in the first sentence of the corpus
```

```python
# TODO: For Binary BOW
cv = CountVectorizer(binary=True,ngram_range=(3,3))  # binary=True means it will only consider the presence or absence of a word, not the frequency,
X = cv.fit_transform(corpus)
X[0].toarray()  # It will display the count of each word in the first sentence of the corpus in binary format (1 if the word is present, 0 if not)
```
